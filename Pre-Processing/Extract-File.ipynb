```python
import pandas as pd
```

```python
df = pd.read_csv("Data/Data_Folder/updated-animal-data-farm-id-70.csv")
df.head()
```

```python
df['id'].unique()
```

```python
df["datetime"] = pd.to_datetime(df["datetime"], errors="coerce")

df = df.sort_values("datetime")

start_date = pd.Timestamp("2025-09-01")
end_date = pd.Timestamp("2025-10-31 23:59:59")  

filtered_df = df[(df["datetime"] >= start_date) & (df["datetime"] <= end_date) & (df["id"] == 6255)]
```

```python
cols_to_drop = ['event', 'date', 'time']  # drop irrevelant columns :)
df_filtered = filtered_df.drop(columns=[col for col in cols_to_drop if col in filtered_df.columns])

output_path = 'Data/Test/Test-animal-id-6255-farm-Id-70.csv' 
df_filtered.to_csv(output_path, index=False)

print(f"File saved to {output_path}")
```
